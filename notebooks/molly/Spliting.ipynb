```python

import pandas as pd

```

```python
#import os
#print(os.getcwd())
data = pd.read_csv(r"C:\Users\mvesk\RiceBootcamp_project1\data_bootcamp_project_1\data\IMDB-Movie-Data.csv")
#data = pd.read_csv('..\data\IMDB-Movie-Data.csv')
data.head()
```

```python
# dropping null value columns to avoid errors 
#data.dropna(inplace = True) 

# new data frame with split value columns 
new = data["Genre"].str.split(",", n = 2, expand = True) 
  
# making seperate first name column from new data frame 
data["Genre1"]= new[0] 
  
# making seperate last name column from new data frame 
data["Genre2"]= new[1]
data["Genre3"]= new[2] 
  
# Dropping old Name columns 
data.drop(columns =["Genre"], inplace = True) 
  
# df display 
data 

```

```python
# new data frame with split value columns 
new2 = data["Actors"].str.split(",", n = 3, expand = True) 
  
# making seperate first name column from new data frame 
data["Actor1"]= new2[0] 
  
# making seperate last name column from new data frame 
data["Actor2"]= new2[1]
data["Actor3"]= new2[2] 
data["Actor4"]= new2[3] 
  
# Dropping old Name columns 
data.drop(columns =["Actors"], inplace = True) 
  
# df display 
data 
```

```python
# fill none value with NAN to avoid errors 
data.fillna("NaN", inplace = True) 
data
```
